```python
from elasticsearch import Elasticsearch
es = Elasticsearch("http://host.docker.internal:9200")
# print(es.info())

# res = es.get(index="my_index", doc_type='my_type', id="my_id5")
hits = es.search(index="my_index", body={"query": {"match_all": {}}})
print("Got %d Hits:" % hits['hits']['total'])
for hit in hits['hits']['hits']:
    print(hit['_source']['url'])

# print(res['_source']['attachment']['content_type'])
# print(res['_source']['url'])
# data = res['_source']['data']
```

```python
import base64
import IPython
import re

# print (dataorig)
for hit in hits['hits']['hits']:
    data = hit['_source']['data']
    content_type = hit['_source']['attachment']['content_type']
    print(content_type)
    if content_type.startswith('text/html'): 
        print(hit['_source']['url'])
        k = content_type.rfind("=")
        dataorig = base64.b64decode(data).decode(content_type[k+1:]) # extract charset
        html = IPython.display.HTML(data=dataorig)
#         display(html)
    elif content_type.startswith('application/pdf'): 
        dataorig = base64.b64decode(data)
        IPython.display.display_pdf(dataorig, raw=True)
    elif content_type.startswith('application/xml'): 
        dataorig = base64.b64decode(data).decode('utf-8')
        xml = IPython.display.Code(data=dataorig, language='xml')
        display(xml)
        removedups = re.sub(' +',' ', hit['_source']['attachment']['content'])
#         filtered = re.sub(r'^$\n', '', removedups, flags=re.MULTILINE)
#         filtered = filter(lambda x:  not re.match(r'^\s*$', x), removedups)
        print('filtered', str(removedups))
    else:
        print('Unknown type', content_type)
```
